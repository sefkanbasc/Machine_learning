```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

dataset = pd.read_csv("C:/Sefkan/UserCarData.csv")
dataset.head()
```

```python
data = dataset[dataset["transmission"]=="Automatic"]
```

```python
x = np.array(data.loc[:,"year"]).reshape(-1,1)
y = np.array(data.loc[:,"km_driven"]).reshape(-1,1)
plt.scatter(x,y)
plt.xlabel("Year")
plt.ylabel("Km_driven")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

linear_regression = LinearRegression()
predict_space = np.linspace(min(x),max(x)).reshape(-1,1)
linear_regression.fit(x,y)

predicted = linear_regression.predict(predict_space)
print("R^2 score: ",linear_regression.score(x,y))
plt.plot(predict_space,predicted,color="red",linewidth=3)
plt.scatter(x,y)
plt.xlabel("Year")
plt.ylabel("Km_driven")
plt.show()
```

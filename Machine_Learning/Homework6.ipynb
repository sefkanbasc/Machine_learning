```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_breast_cancer
breast = load_breast_cancer()
breast_data = breast.data
breast_data.shape
```

```python
breast_labels = breast.target
breast_labels.shape
```

```python
labels = np.reshape(breast_labels,(569,1))
final_breast_data = np.concatenate([breast_data,labels],axis=1)
final_breast_data.shape
```

```python
breast_dataset = pd.DataFrame(final_breast_data)
features = breast.feature_names
features
```

```python
features_labels = np.append(features,"label")
breast_dataset.columns = features_labels
breast_dataset.head()
```

```python
breast_dataset['label'].replace(0, 'Benign',inplace=True)
breast_dataset['label'].replace(1, 'Malignant',inplace=True)

breast_dataset.tail()
```

```python
x = breast_dataset.iloc[:,[2,3]].values
y = breast_dataset.iloc[:,4].values
```

```python
from sklearn.model_selection import train_test_split
x_train, y_train, x_test, y_test = train_test_split(x,y, test_size=0.25,random_state=0)
```

```python
print("Training Data Shape:",x_train.shape)
print("Test Data Shape:",x_test.shape)
```

```python
print("Training Data Shape:",y_train.shape)
print("Test Data Shape:",y_test.shape)
```

```python
classes = np.unique(y_train)
nClasses = len(classes)
print("Total number of outputs:",nClasses)
print("Output classes:",classes)
```

```python
from sklearn.preprocessing import StandardScaler
x = breast_dataset.loc[:,features].values
x = StandardScaler().fit_transform(x)
x.shape
```

```python
np.mean(x),np.std(x)
```

```python
feat_cols = ["feature"+str(i) for i in range(x.shape[1])]
```

```python
normalised_breast = pd.DataFrame(x,columns=feat_cols)
normalised_breast.tail()
```

```python
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=2)
principalComponent_breast = pca_breast.fit_transform(x)
```

```python
principal_breast_Df = pd.DataFrame(data = principalComponent_breast
             , columns = ['principal component 1', 'principal component 2'])
principal_breast_Df.tail()
```

```python
print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
```
